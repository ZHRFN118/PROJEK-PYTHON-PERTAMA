KALKULATOR MUDAH

```python
import math

def show_menu():
    print("\n" + "="*40)
    print("         SMART PYTHON CALCULATOR        ")
    print("="*40)
    print("Type your math equation and press Enter.")
    print("Supported: +, -, *, /, **, ( ), pi, e")
    print("Type 'exit' to quit.")
    print("-"*40)

def smart_calculator():
    # Allowed math terms for security/safety
    allowed_words = {
        'pi': math.pi,
        'e': math.e,
    }

    show_menu()

    while True:
        user_input = input("\nCalculate > ").strip()

        if user_input.lower() == 'exit':
            print("Goodbye! Thanks for using the calculator.")
            break

        if not user_input:
            continue

        try:
            # eval evaluates the string as code.
            # We pass allowed_words to let users use 'pi' or 'e' safely.
            result = eval(user_input, {"__builtins__": None}, allowed_words)

            # Format result nicely: hide decimal if it's a whole number
            if isinstance(result, float) and result.is_integer():
                result = int(result)

            print(f"Result: {result}")

        except ZeroDivisionError:
            print("Error: You cannot divide by zero!")
        except (SyntaxError, NameError, TypeError):
            print("Error: Invalid equation. Use numbers and standard operators (+, -, *, /).")
        except Exception as e:
            print(f"Error: Could not calculate. ({e})")

if __name__ == "__main__":
    smart_calculator()
```
