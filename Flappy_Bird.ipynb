```python
from IPython.display import display, HTML

# Ambil kod penuh dari sini sehingga ke hujung bawah sekali
flappy_colab_html = """
<div style="text-align: center; background-color: #2c3e50; padding: 15px; border-radius: 12px; width: 430px; margin: auto; box-shadow: 0px 4px 15px rgba(0,0,0,0.3);">
    <h2 style="color: #fff; font-family: 'Courier New', Courier, monospace; margin: 0 0 5px 0; text-transform: uppercase; letter-spacing: 2px;">Flappy Bird</h2>

    <div style="display: flex; justify-content: space-around; color: #fff; font-family: 'Courier New', Courier, monospace; font-size: 18px; margin-bottom: 10px;">
        <div style="color: #ffcc00;">Skor: <span id="score">0</span></div>
        <div style="color: #00ffcc;">🏆 Terbaik: <span id="highscore">0</span></div>
    </div>

    <canvas id="gameCanvas" width="400" height="480" style="border: 4px solid #fff; background-color: #70c5ce; border-radius: 5px; cursor: pointer;"></canvas>
    <p style="color: #bdc3c7; font-family: sans-serif; font-size: 12px; margin-top: 8px; line-height: 1.4;">
        Tekan <b>SPACEBAR</b> pada keyboard atau <b>KLIK/KETIK</b> pada skrin untuk melompat.<br>
        <i>Highscore akan disimpan walaupun anda run semula kod ini!</i>
    </p>
</div>

<script>
    const canvas = document.getElementById("gameCanvas");
    const ctx = canvas.getContext("2d");
    const scoreTxt = document.getElementById("score");
    const highscoreTxt = document.getElementById("highscore");

    let highscore = localStorage.getItem("flappy_highscore") || 0;
    highscoreTxt.innerText = highscore;

    let birdY = 200;
    let velocity = 0;
    const gravity = 0.35;
    const jump = -6.0;
    const birdX = 70;
    const birdRadius = 12;

    let pipes = [];
    const pipeWidth = 55;
    const pipeGap = 135;
    const pipeSpeed = 2.2;

    let score = 0;
    let gameActive = true;
    let loopId;

    function initGame() {
        birdY = 200;
        velocity = 0;
        pipes = [{ x: canvas.width, top: 140, bottom: 140 + pipeGap, passed: false }];
        score = 0;
        scoreTxt.innerText = score;
        gameActive = true;
        if(loopId) cancelAnimationFrame(loopId);
        gameLoop();
    }

    function addPipe() {
        const minH = 40;
        const maxH = canvas.height - pipeGap - minH;
        const topH = Math.floor(Math.random() * (maxH - minH + 1)) + minH;
        pipes.push({ x: canvas.width, top: topH, bottom: topH + pipeGap, passed: false });
    }

    function gameLoop() {
        if (!gameActive) return;

        velocity += gravity;
        birdY += velocity;

        if (birdY + birdRadius >= canvas.height || birdY - birdRadius <= 0) {
            triggerGameOver();
            return;
        }

        ctx.fillStyle = "#70c5ce";
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        for (let i = pipes.length - 1; i >= 0; i--) {
            pipes[i].x -= pipeSpeed;

            ctx.fillStyle = "#2ecc71";
            ctx.fillRect(pipes[i].x, 0, pipeWidth, pipes[i].top);
            ctx.strokeStyle = "#27ae60";
            ctx.lineWidth = 3;
            ctx.strokeRect(pipes[i].x, -5, pipeWidth, pipes[i].top + 5);

            ctx.fillRect(pipes[i].x, pipes[i].bottom, pipeWidth, canvas.height - pipes[i].bottom);
            ctx.strokeRect(pipes[i].x, pipes[i].bottom, pipeWidth, canvas.height - pipes[i].bottom + 5);

            if (birdX + birdRadius > pipes[i].x && birdX - birdRadius < pipes[i].x + pipeWidth &&
                (birdY - birdRadius < pipes[i].top || birdY + birdRadius > pipes[i].bottom)) {
                triggerGameOver();
                return;
            }

            if (!pipes[i].passed && pipes[i].x + pipeWidth < birdX) {
                pipes[i].passed = true;
                score++;
                scoreTxt.innerText = score;

                if (score > highscore) {
                    highscore = score;
                    highscoreTxt.innerText = highscore;
                    localStorage.setItem("flappy_highscore", highscore);
                }
            }

            if (pipes[i].x + pipeWidth < 0) pipes.splice(i, 1);
        }

        if (pipes.length === 0 || pipes[pipes.length - 1].x < canvas.width - 200) {
            addPipe();
        }

        ctx.beginPath();
        ctx.arc(birdX, birdY, birdRadius, 0, Math.PI * 2);
        ctx.fillStyle = "#f1c40f";
        ctx.fill();
        ctx.strokeStyle = "#d35400";
        ctx.lineWidth = 2;
        ctx.stroke();

        ctx.beginPath();
        ctx.arc(birdX - 4, birdY + 2, 5, 0, Math.PI * 2);
        ctx.fillStyle = "#f39c12";
        ctx.fill();

        ctx.beginPath();
        ctx.arc(birdX + 4, birdY - 3, 3, 0, Math.PI * 2);
        ctx.fillStyle = "#fff";
        ctx.fill();
        ctx.beginPath();
        ctx.arc(birdX + 5, birdY - 3, 1.5, 0, Math.PI * 2);
        ctx.fillStyle = "#000";
        ctx.fill();

        ctx.fillStyle = "#e67e22";
        ctx.beginPath();
        ctx.moveTo(birdX + birdRadius - 2, birdY - 2);
        ctx.lineTo(birdX + birdRadius + 5, birdY + 1);
        ctx.lineTo(birdX + birdRadius - 2, birdY + 4);
        ctx.fill();

        loopId = requestAnimationFrame(gameLoop);
    }

    function triggerGameOver() {
        gameActive = false;
        ctx.fillStyle = "rgba(0, 0, 0, 0.6)";
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        ctx.fillStyle = "#fff";
        ctx.font = "bold 26px Courier New";
        ctx.textAlign = "center";
        ctx.fillText("PERMAINAN TAMAT", canvas.width / 2, canvas.height / 2 - 15);

        ctx.fillStyle = "#ffcc00";
        ctx.font = "16px Courier New";
        ctx.fillText("Klik Skrin / Tekan R untuk Main Semula", canvas.width / 2, canvas.height / 2 + 30);
    }

    window.addEventListener("keydown", e => {
        if (e.code === "Space") {
            e.preventDefault();
            if (gameActive) velocity = jump;
        }
        if (e.key === "r" || e.key === "R") {
            if (!gameActive) initGame();
        }
    });

    canvas.addEventListener("click", () => {
        if (gameActive) {
            velocity = jump;
        } else {
            initGame();
        }
    });

    initGame();
</script>
"""

# Paparkan ke skrin Google Colab selepas strings ditutup rapat
display(HTML(flappy_colab_html))

import urllib.request
from IPython.display import display, HTML

# 1. Memuat turun kod permainan penuh yang lengkap dari server luar
urllib.request.urlretrieve("https://pastebin.com/raw/p13e9a4J", "flappy.html")

# 2. Membaca dan memaparkan permainan di Google Colab
with open("flappy.html", "r", encoding="utf-8") as f:
    display(HTML(f.read()))
```
